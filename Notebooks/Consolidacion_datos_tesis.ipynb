```python
def tratamiento_datos(df_colegio):
    data_dapurada2=df_colegio[['ESTU_GENERO','PERIODO','ESTU_TIENEETNIA',
           'FAMI_ESTRATOVIVIENDA', 'FAMI_PERSONASHOGAR',
           'FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE',
           'FAMI_TIENEINTERNET', 'FAMI_TIENECOMPUTADOR', 
                               #'FAMI_SITUACIONECONOMICA',
           #'ESTU_DEDICACIONLECTURADIARIA', 'ESTU_DEDICACIONINTERNET',
           'COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_NOMBRE_ESTABLECIMIENTO',
           'COLE_GENERO', 'COLE_NATURALEZA', 'COLE_CALENDARIO', 'COLE_BILINGUE',
           'COLE_CARACTER', 'COLE_COD_DANE_SEDE', 'COLE_NOMBRE_SEDE',
           'COLE_SEDE_PRINCIPAL', 'COLE_AREA_UBICACION',
           'COLE_COD_MCPIO_UBICACION', 'COLE_MCPIO_UBICACION',
           'COLE_COD_DEPTO_UBICACION', 'COLE_DEPTO_UBICACION','PUNT_GLOBAL','PERCENTIL_GLOBAL']]
    data_dapurada2["CANTIDAD_ESTUDIANTES"]=1
    datos_dumis2=pd.get_dummies(data_dapurada2 ,columns=["ESTU_GENERO","ESTU_TIENEETNIA","FAMI_ESTRATOVIVIENDA",
                                                         "FAMI_PERSONASHOGAR", 'FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE',
                                                         'FAMI_TIENEINTERNET', 
                                                         'FAMI_TIENECOMPUTADOR'
                                                         #,"FAMI_SITUACIONECONOMICA","ESTU_DEDICACIONLECTURADIARIA"
                                                         #,"ESTU_DEDICACIONINTERNET"
                                                        ])
    datos_dumis2.head()
    data_max=datos_dumis2.groupby(['PERIODO','COLE_COD_DANE_ESTABLECIMIENTO', 'COLE_NOMBRE_ESTABLECIMIENTO',
           'COLE_GENERO', 'COLE_NATURALEZA', 'COLE_CALENDARIO', 'COLE_BILINGUE',
           'COLE_CARACTER', 'COLE_COD_DANE_SEDE', 'COLE_NOMBRE_SEDE',
           'COLE_SEDE_PRINCIPAL', 'COLE_AREA_UBICACION', 
           'COLE_COD_MCPIO_UBICACION', 'COLE_MCPIO_UBICACION',
           'COLE_COD_DEPTO_UBICACION', 'COLE_DEPTO_UBICACION'], as_index = False).agg({
        'PUNT_GLOBAL':['mean', 'median','std'],
                      'ESTU_GENERO_F':'sum',
                        'ESTU_GENERO_M':'sum',
                        'ESTU_TIENEETNIA_No':'sum',
                        'ESTU_TIENEETNIA_Si':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 1':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 2':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 3':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 4':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 5':'sum',
                        'FAMI_ESTRATOVIVIENDA_Estrato 6':'sum',
                        #'FAMI_ESTRATOVIVIENDA_Sin Estrato':'sum',
                        #'FAMI_PERSONASHOGAR_1 a 2':'sum',
                        #'FAMI_PERSONASHOGAR_3 a 4':'sum',
                        #'FAMI_PERSONASHOGAR_5 a 6':'sum',
                        #'FAMI_PERSONASHOGAR_7 a 8':'sum',
                        #'FAMI_PERSONASHOGAR_9 o más':'sum',
                        'FAMI_EDUCACIONPADRE_Educación profesional completa':'sum',
                        'FAMI_EDUCACIONPADRE_Educación profesional incompleta':'sum',
                        'FAMI_EDUCACIONPADRE_Ninguno':'sum',
                        #'FAMI_EDUCACIONPADRE_No Aplica':'sum',
                        'FAMI_EDUCACIONPADRE_No sabe':'sum',
                        'FAMI_EDUCACIONPADRE_Postgrado':'sum',
                        'FAMI_EDUCACIONPADRE_Primaria completa':'sum',
                        'FAMI_EDUCACIONPADRE_Primaria incompleta':'sum',
                        'FAMI_EDUCACIONPADRE_Secundaria (Bachillerato) completa':'sum',
                        'FAMI_EDUCACIONPADRE_Secundaria (Bachillerato) incompleta':'sum',
                        'FAMI_EDUCACIONPADRE_Técnica o tecnológica completa':'sum',
                        'FAMI_EDUCACIONPADRE_Técnica o tecnológica incompleta':'sum',
                        'FAMI_EDUCACIONMADRE_Educación profesional completa':'sum',
                        'FAMI_EDUCACIONMADRE_Educación profesional incompleta':'sum',
                        'FAMI_EDUCACIONMADRE_Ninguno':'sum',
                        #'FAMI_EDUCACIONMADRE_No Aplica':'sum',
                        'FAMI_EDUCACIONMADRE_No sabe':'sum',
                        'FAMI_EDUCACIONMADRE_Postgrado':'sum',
                        'FAMI_EDUCACIONMADRE_Primaria completa':'sum',
                        'FAMI_EDUCACIONMADRE_Primaria incompleta':'sum',
                        'FAMI_EDUCACIONMADRE_Secundaria (Bachillerato) completa':'sum',
                        'FAMI_EDUCACIONMADRE_Secundaria (Bachillerato) incompleta':'sum',
                        'FAMI_EDUCACIONMADRE_Técnica o tecnológica completa':'sum',
                        'FAMI_EDUCACIONMADRE_Técnica o tecnológica incompleta':'sum',
                        'FAMI_TIENEINTERNET_No':'sum',
                        'FAMI_TIENEINTERNET_Si':'sum',
                        'FAMI_TIENECOMPUTADOR_No':'sum',
                        'FAMI_TIENECOMPUTADOR_Si':'sum',
                        #'FAMI_SITUACIONECONOMICA_Igual':'sum',
                        #'FAMI_SITUACIONECONOMICA_Mejor':'sum',
                        #'FAMI_SITUACIONECONOMICA_Peor':'sum',
                        #'ESTU_DEDICACIONLECTURADIARIA_30 minutos o menos':'sum',
                        #'ESTU_DEDICACIONLECTURADIARIA_Entre 1 y 2 horas':'sum',
                        #'ESTU_DEDICACIONLECTURADIARIA_Entre 30 y 60 minutos':'sum',
                        #'ESTU_DEDICACIONLECTURADIARIA_Más de 2 horas':'sum',
                        #'ESTU_DEDICACIONLECTURADIARIA_No leo por entretenimiento':'sum',
                        #'ESTU_DEDICACIONINTERNET_30 minutos o menos':'sum',
                        #'ESTU_DEDICACIONINTERNET_Entre 1 y 3 horas':'sum',
                        #'ESTU_DEDICACIONINTERNET_Entre 30 y 60 minutos':'sum',
                        #'ESTU_DEDICACIONINTERNET_Más de 3 horas':'sum',
                        #'ESTU_DEDICACIONINTERNET_No Navega Internet':'sum',
                        "CANTIDAD_ESTUDIANTES":'sum'})
    return data_max
```
